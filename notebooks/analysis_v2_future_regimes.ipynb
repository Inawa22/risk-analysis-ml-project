We start by loading market data (S&P 500, 2005–date)

```python
import yfinance as yf
import pandas as pd
import numpy as np

# Download S&P 500 data
sp500 = yf.download("^GSPC", start="2005-01-01")

# Columns are MultiIndex, so we pick the level that contains OHLCV field names
field_names = {"Open", "High", "Low", "Close", "Adj Close", "Volume"}

lvl0 = set(map(str, sp500.columns.get_level_values(0)))
lvl1 = set(map(str, sp500.columns.get_level_values(1)))

# If level 0 looks like fields, keep level 0; else keep level 1
if len(field_names.intersection(lvl0)) >= 3:
    sp500.columns = sp500.columns.get_level_values(0)
elif len(field_names.intersection(lvl1)) >= 3:
    sp500.columns = sp500.columns.get_level_values(1)
else:
    # last resort: flatten by joining both levels
    sp500.columns = [f"{a}_{b}" for a, b in sp500.columns]

# Keep standard OHLCV
sp500 = sp500.loc[:, ["Open", "High", "Low", "Close", "Volume"]].copy()
sp500.dropna(inplace=True)

print("INDEX TYPE:", type(sp500.index))
print("COLUMNS TYPE:", type(sp500.columns))
print("COLUMNS:", sp500.columns.tolist()[:10])
sp500.tail()
```

##Feature engineering —
To define the features, we separate market and macro features

The market features are computed below

```python
# Daily returns
sp500['return'] = sp500['Close'].pct_change()

# Rolling volatility
for window in [5, 10, 20]:
    sp500[f'vol_{window}d'] = sp500['return'].rolling(window).std()

# Rolling volume
sp500['vol_volume_20d'] = sp500['Volume'].rolling(20).mean()

# High–low range
sp500['hl_range'] = (sp500['High'] - sp500['Low']) / sp500['Close']

# Momentum
sp500['momentum_5d'] = sp500['return'].rolling(5).sum()
sp500['momentum_10d'] = sp500['return'].rolling(10).sum()

sp500.dropna(inplace=True)
```

Define the target variable (risk regime)

```python
# 20-day realised volatility
sp500['realised_vol_20d'] = sp500['return'].rolling(20).std()

# High-volatility threshold (top 25%)
threshold = sp500['realised_vol_20d'].quantile(0.75)

# Binary risk regime
sp500['high_vol_regime'] = (sp500['realised_vol_20d'] >= threshold).astype(int)

sp500[['realised_vol_20d', 'high_vol_regime']].tail()
```

Macro series are often monthly, while market data is daily. The model requires one row per trading day. So, we forward-fill.

```python
daily_index = sp500.index

macro_daily = macro.reindex(daily_index).ffill()
macro_daily.tail()
```

We lag monthly macro series by ~1 month worth of trading days (approx 22) and daily VIX by 1 day (yesterday's VIX known at today's open)

```python
# Conservative anti-leakage lagging:

macro_daily_lagged = macro_daily.copy()

# Lag monthly-ish series by 22 trading days
for col in ["fed_funds", "cpi", "unemp"]:
    macro_daily_lagged[col] = macro_daily_lagged[col].shift(22)

# Lag VIX by 1 day
macro_daily_lagged["vix_fred"] = macro_daily_lagged["vix_fred"].shift(1)

macro_daily_lagged.tail()
```

We then join both frames together and drop rows with NA introduced by rolling windows and lagging

```python
df = sp500.join(macro_daily_lagged, how="inner")

df = df.dropna()

df.head()
```

```python
# Check target balance
print(df["high_vol_regime"].value_counts(normalize=True))

# Ensure no obvious missingness
df.isna().mean().sort_values(ascending=False).head(10)
```

Time-series train / validation / test split

Important to note: we do not shuffle because this is time-series data, hence it has time order and memory

```python
from datetime import datetime

target = "high_vol_regime"

feature_cols = [
    # market features
    "return", "vol_5d", "vol_10d", "vol_20d",
    "vol_volume_20d", "hl_range",
    "momentum_5d", "momentum_10d",
    # macro features
    "fed_funds", "cpi", "unemp", "vix_fred",
]

train_end = "2017-12-31"
val_end = "2019-12-31"

train = df.loc[:train_end]
val   = df.loc[train_end:val_end]
test  = df.loc[val_end:]

X_train, y_train = train[feature_cols], train[target]
X_val,   y_val   = val[feature_cols],   val[target]
X_test,  y_test  = test[feature_cols],  test[target]
```

Our initial baseline is a Logistic Regression classifier.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

logit = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=1000, class_weight="balanced"))
])

logit.fit(X_train, y_train)
```

To capture non-linear interactions, we then train a Gradient Boosting classifier.

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=3,
    random_state=42
)

gb.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score, classification_report

def evaluate(model, X, y, name="Model"):
    probs = model.predict_proba(X)[:, 1]
    preds = (probs >= 0.5).astype(int)

    print(f"\n{name}")
    print("ROC-AUC:", roc_auc_score(y, probs))
    print(classification_report(y, preds))

evaluate(logit, X_val, y_val, "Logistic (Validation)")
evaluate(gb,    X_val, y_val, "GB (Validation)")
```

```python
covid_period = (X_test.index >= "2020-02-01") & (X_test.index <= "2020-06-30")

evaluate(logit, X_test.loc[covid_period], y_test.loc[covid_period], "Logistic (COVID)")
evaluate(gb,    X_test.loc[covid_period], y_test.loc[covid_period], "GB (COVID)")
```

The realised volatility feature dominates because the target was derived from it(it is perfectly correlated to the target). We identify this as a form of target leakage by construction and reframethe task to predict future volatility regimes, ensuring the model relied on leading indicators rather than tautological signals.

```python
import numpy as np
import pandas as pd

# 1) Define threshold on TRAIN ONLY (prevents leakage)
train_end = "2017-12-31"   # same as earlier split
q = 0.75                   # top 25% = "high vol"

threshold_train = df.loc[:train_end, "realised_vol_20d"].quantile(q)

# 2) Create FUTURE target: will volatility be high H days ahead?
H = 5  # forecast horizon in trading days
df["high_vol_regime_future"] = (df["realised_vol_20d"].shift(-H) >= threshold_train).astype(int)

# 3) Drop rows that can't have a future label (last H rows) and any NAs
df_future = df.dropna(subset=feature_cols + ["high_vol_regime_future"]).copy()

# Optional sanity check: Target prevalence looks reasonable ^ Threshold is computed on training data only
print("Target positive rate:", df_future["high_vol_regime_future"].mean())
print("Train threshold (q=0.75) from train only:", threshold_train)

# 5) Time-based split (same dates, but using FUTURE target)
target_future = "high_vol_regime_future"

train = df_future.loc[:train_end]
val_end = "2019-12-31"
val   = df_future.loc[train_end:val_end]
test  = df_future.loc[val_end:]

X_train, y_train = train[feature_cols], train[target_future]
X_val,   y_val   = val[feature_cols],   val[target_future]
X_test,  y_test  = test[feature_cols],  test[target_future]

# 6) Re-train models (baseline + stronger)
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import roc_auc_score, classification_report

logit = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=1000, class_weight="balanced"))
])
logit.fit(X_train, y_train)

gb = GradientBoostingClassifier(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=3,
    random_state=42
)
gb.fit(X_train, y_train)

# 7) Evaluation helper
def evaluate(model, X, y, name="Model", threshold=0.5):
    probs = model.predict_proba(X)[:, 1]
    preds = (probs >= threshold).astype(int)
    print(f"\n{name} @ threshold={threshold}")
    print("ROC-AUC:", roc_auc_score(y, probs))
    print(classification_report(y, preds))

# Validate performance
evaluate(logit, X_val, y_val, "Logistic (Validation)")
evaluate(gb,    X_val, y_val, "GB (Validation)")

# Stress test (COVID)
covid_mask = (X_test.index >= "2020-02-01") & (X_test.index <= "2020-06-30")
evaluate(logit, X_test.loc[covid_mask], y_test.loc[covid_mask], "Logistic (COVID)")
evaluate(gb,    X_test.loc[covid_mask], y_test.loc[covid_mask], "GB (COVID)")

# 8) Feature importance (GB)
feat_imp = pd.Series(gb.feature_importances_, index=feature_cols).sort_values(ascending=False)
print("\nTop feature importances (GB):")
print(feat_imp.head(15))

```
